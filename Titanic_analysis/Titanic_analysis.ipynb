```python
#This plot Represent counts of passengers in each age group 

sns.displot(csv.Age, color='cyan', rug=True, kde=True)
```

```python
# it helps in comparing Age distribution for survivors and non-survivors, separated by Sex (hue).
sns.set(style='whitegrid')
sns.boxplot(x='Survived', y='Age', data=csv,palette='spring', hue='Sex')
```

```python
#this plot  visually compares age distribution across passenger classes by gender

sns.scatterplot(x='Pclass', y='Age', hue='Sex', style='Sex',data=csv, palette='ocean',size='Sex',sizes=(150,100),markers={'female':'d','male':'s'})
```

```python
#It helps in detecting patterns, correlations, and distribution differences between survivors and non-survivors.
sns.pairplot(csv[["Age", "Fare", "SibSp", "Parch", "Survived"]],
             hue="Survived", diag_kind="kde")
```

```python
# Itâ€™s a color-coded map showing how strongly the Titanic datasetâ€™s numeric features are related to each other.
# Colors show correlation strength:
# Red â†’ Positive correlation
# Blue â†’ Negative correlation
# White â†’ Near zero correlation
corr = csv[["Age", "Fare", "SibSp", "Parch", "Survived", "Pclass"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")

```
